## 1. Extracting all Uniprot ID - RHEA reaction ID pairs from the downloaded file from UniProt:

```python
df_uniprot = pd.read_excel(join("..", "..", "data", "Uniprot", "uniprot-reviewed yes_transport.xlsx"))
df_uniprot
```

```python
start_rhea = "Xref=Rhea:"

def get_reaction_info(activity):
    reaction = activity.split(";")[0]
    reaction = reaction.replace("Reaction=", "")
    
    start_pos = activity.find(start_rhea)
    if start_pos != -1:
        rhea_ID = activity[start_pos + len(start_rhea):]
        rhea_ID = rhea_ID.split(",")[0]
    else:
        rhea_ID = np.nan
        
    return(rhea_ID, reaction)


def get_activtities(activities):
    activities = activities.split("CATALYTIC ACTIVITY: ")
    return(activities[1:])
```

```python
df_reactions = pd.DataFrame(columns = ["Uniprot ID", "Reaction", "RHEA ID", "GO"])

for ind in df_uniprot.index:
    if ind % 1000 ==0:
        print(ind)
    activities = df_uniprot["Catalytic activity"][ind]
    if not pd.isnull(activities):
        activities = get_activtities(activities = activities)
        for activity in activities:
            if "(out)" in activity and "(in)" in activity:
                rhea_ID, reaction = get_reaction_info(activity)
                df_reactions = df_reactions.append({"Uniprot ID" : df_uniprot["Entry"][ind],
                                                    "Reaction" : reaction, 
                                                    "RHEA ID" : rhea_ID,
                                                   "GO" : df_uniprot["Gene ontology (molecular function)"][ind]
                                                   },
                                                   ignore_index=True)
                
                
                
print("Number of different UniProt IDs : %s \n\
Number of different RHEA reaction IDs: %s" % (len(set(list(df_reactions["Uniprot ID"]))), len(set(list(df_reactions["RHEA ID"])))))                

df_reactions
```

### 2. Mapping RHEA reaction IDs to substrate IDs:

### (a) Mapping RHEA reaction IDs to transported substrates:

```python
def extract_RHEA_ID_and_CHEBI_IDs(entry):
    RHEA_ID = entry[0][len("ENTRY"): -1]
    RHEA_ID = RHEA_ID.split(" ")[-1]
    CHEBI_IDs = entry[2][len("EQUATION"): -1]
    CHEBI_IDs = CHEBI_IDs[CHEBI_IDs.index("CHEBI"):]
    
    sub_names = entry[1][len("DEFINITION"): -1]
    while sub_names[0] == " ":
        sub_names = sub_names[1:]
    
    return(RHEA_ID, CHEBI_IDs, sub_names)

def get_substrate_IDs(IDs):
    IDs = IDs.split(" = ")[0]
    IDs = IDs.split(" => ")[0]
    IDs = IDs.split(" <=> ")[0]
    IDs = IDs.replace(" + ", ";")
    IDs = IDs.split(";")
    return([ID.split(" ")[-1] for ID in IDs])

def get_substrate_names(names):
    names = names.split(" = ")[0]
    names = names.split(" => ")[0]
    names = names.split(" <=> ")[0]
    names = names.replace(" + ", ";")
    names = names.split(";")
    return([ID.split(" ")[-1] for ID in names])

def get_acess_array(names):
    []
```

```python
df_RHEA = pd.DataFrame(columns = ["RHEA ID", "CHEBI_ID_list"])

file1 = open(join("..", "..", "data", "reactions", "rhea-reactions.txt"), 'r')
Lines = file1.readlines()

while True:
    try:
        end = Lines.index('///\n')
        entry = Lines[:end]
        RHEA_ID, CHEBI_IDs, sub_names = extract_RHEA_ID_and_CHEBI_IDs(entry)
        CHEBI_ID_list, sub_name_list = get_substrate_IDs(IDs = CHEBI_IDs), get_substrate_names(names = sub_names)
        access_array = [True if "(out)" in sub or "(in)" in sub else False for sub in sub_name_list ]
        CHEBI_ID_list = list(np.array(CHEBI_ID_list)[access_array])
        Lines = Lines[end+1:]
        df_RHEA = df_RHEA.append({"RHEA ID" : RHEA_ID, "CHEBI_ID_list" : CHEBI_ID_list}, ignore_index = True)
    except ValueError:
        break
        
df_RHEA["RHEA ID"] = [float(ID.split(":")[-1]) for ID in df_RHEA["RHEA ID"]]
df_RHEA.to_pickle(join("..", "..", "data", "reactions", "RHEA_reaction_df.pkl"))
df_RHEA
```

```python
def get_CHEBI_IDs_from_RHEA_ID(rhea_ID):
    rhea_ID = float(rhea_ID.replace("RHEA:", ""))
    try:
        CHEBI_IDs = list(df_RHEA["CHEBI_ID_list"].loc[df_RHEA["RHEA ID"] == rhea_ID])[0]
    except IndexError:
        CHEBI_IDs = []
    return(CHEBI_IDs)
```

```python
df_prot_sub = pd.DataFrame(columns = ["Uniprot ID", "CHEBI ID"])

for ind in df_reactions.index:
    rhea_ID = df_reactions["RHEA ID"][ind]
    if not pd.isnull(rhea_ID):
        CHEBI_IDs = get_CHEBI_IDs_from_RHEA_ID(rhea_ID = rhea_ID)
        UID = df_reactions["Uniprot ID"][ind]
        for CHEBI_ID in CHEBI_IDs:
            df_prot_sub = df_prot_sub.append({"Uniprot ID" : UID, "CHEBI ID" : CHEBI_ID}, ignore_index=True)
df_prot_sub.drop_duplicates(inplace = True)
df_prot_sub
```

```python
df_prot_sub.rename(columns = {"CHEBI ID" : "molecule ID"}, inplace = True)
```
